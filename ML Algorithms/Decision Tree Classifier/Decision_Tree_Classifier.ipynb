## **Importing Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
```

```python
data_set = pd.read_csv("dataset01.csv")
data = pd.DataFrame(data_set)
print(data)
```

## **Feature Extraction**

```python
lb = preprocessing.LabelEncoder()
features = list(zip(data["Height"],data["Weight"],data["Foot_Size"]))
label = lb.fit_transform(data["Gender"])
```

## **Fitting Model**

```python
classifier = DecisionTreeClassifier()
fitModel = classifier.fit(features, label)
```

## **Model Predicting with user provided data**

```python
height = input("Enter your Height (in ft): ")
weight = input("Enter your weight (in Lbs): ")
foot_size = input("Enter your foot size: ")
new_data = {"Height":[height],"Weight":[weight],"Foot_Size":[foot_size]}
new_Entry = pd.DataFrame(new_data)
predictions = int(fitModel.predict(new_Entry))
```

```python
def predictionCheck(prediction):
  flag = 0
  isTrue = input("Am I correct? yes or no: ")
  if prediction == 1 and isTrue == "yes":
    print("That's Great! AI is future indeed")
    new_Entry["Gender"] = "male"
  elif prediction == 0 and isTrue == "yes":
    print("That's Great! AI is future indeed")
    new_Entry["Gender"] = "female"
  elif prediction == 1 and isTrue == "no":
    print("Oh! nevermind. I am still learning")
    new_Entry["Gender"] = "female"
  elif prediction == 0 and isTrue == "no":
    print("Oh! nevermind. I am still learning")
    new_Entry["Gender"] = "male"
   
```

## **Appending new data in dataset**

```python
if predictions==1:
  print("I assume that you are male")
  predictionCheck(predictions)
else:
  print("I assume that you are a female")
  predictionCheck(predictions)
data = data.append(new_Entry,ignore_index=True)
data.to_csv("dataset01.csv",index= False)
```
