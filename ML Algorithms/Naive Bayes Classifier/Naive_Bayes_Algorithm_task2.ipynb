# **Solution**

## **Importing Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

## **Data Framing and Labelling**

```python
dataset_file = pd.read_csv("dataset02.csv")
data = pd.DataFrame(dataset_file)
print(data)

le = preprocessing.LabelEncoder()
age = le.fit_transform(data["age"])
income = le.fit_transform(data["income"])
student = le.fit_transform(data["student"])
credit_rating= le.fit_transform(data["credit_rating"])
buys_computer = le.fit_transform(data["buys_computer"])
```

## **Feature Extraction**

```python
features = list(zip(age,income,student,credit_rating))
label = buys_computer
feature_train, feature_test, label_train, label_test = train_test_split(features,label)
```

## **Training Model**

```python
model = GaussianNB()
fitModel = model.fit(feature_train,label_train)
prediction= fitModel.predict(feature_test)
print("our model accuracy score is :",accuracy_score(label_test,prediction))
print("confusion matrix: \n",confusion_matrix(label_test,prediction));
```
